```python
# A simple calculator for basic math operations including power
num1 = float(input("Enter first number: "))
op = input("Enter operator (+, -, *, /, **): ")
num2 = float(input("Enter second number: "))

if op == '+': print(num1 + num2)
elif op == '-': print(num1 - num2)
elif op == '*': print(num1 * num2)
elif op == '/': print(num1 / num2)
elif op == '**': print(num1 ** num2)
else: print("Invalid Operator")
```

```python
# Checks if any item in the list is a number using a loop
my_list = ["test", 25, "apple", 10.5]
for x in my_list:
    if isinstance(x, (int, float)):
        print(f"Numeric value found: {x}")
```

```python
# Adds a new key-value pair to an existing dictionary
my_dict = {"name": "Asad", "age": 20}
my_dict["city"] = "Karachi"
print(my_dict)
```

```python
# Sums all values in a dictionary that are numbers
data = {"v1": 10, "v2": "abc", "v3": 20}
total = sum(v for v in data.values() if isinstance(v, (int, float)))
print(f"Total Sum: {total}")
```

```python
# Finds and prints duplicate items from a list
nums = [1, 2, 3, 2, 1, 4, 5]
duplicates = set([x for x in nums if nums.count(x) > 1])
print(f"Duplicates: {duplicates}")
```

```python
# Checks if a specific key is already present in the dictionary
info = {"id": 1, "status": "active"}
key = "id"
if key in info:
    print(f"Key '{key}' exists!")
else:
    print("Key not found.")
```
